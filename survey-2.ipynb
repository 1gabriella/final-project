```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'age' to numeric
df["age_numeric"] = pd.to_numeric(df["age"], errors="coerce")

#  Filter out rows 
df_clean = df.dropna(subset=["age_numeric", "different_time"]).copy()

counts = df_clean.groupby("different_time").size()
avg_age = df_clean.groupby("different_time")["age_numeric"].mean()


category_order = [
    "no", 
    "sometimes", 
    "yes", 
    "60s-70s", 
    "90s", 
    "00s", 
    "before phones", 
    "like jesus"
]



# Plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Left axis: number of respondentS
ax1.plot(
    category_order, 
    counts, 
    color="blue", 
    marker="o", 
    label="Number of respondents"
)
ax1.set_xlabel("Time Category")
ax1.set_ylabel("Number of Respondents", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

#  Right axis: average age 
ax2 = ax1.twinx()
ax2.plot(
    category_order, 
    avg_age, 
    color="red", 
    marker="s", 
    label="Average Age"
)
ax2.set_ylabel("Average Age", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

plt.title("How Many Want to Be Alive in Each Time & Their Average Age")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python



romantic_counts = df["romanticize_or_sad[]"][df["romanticize_or_sad[]"] != "N/A"].value_counts()


labels = romantic_counts.index
values = romantic_counts.values

# the figure and axes
fig, ax = plt.subplots(figsize=(8, 8))

#donut chart 
wedges, texts, autotexts = ax.pie(
    values,
    labels=None,          # We'll handle legend externally
    autopct='%1.0f%%',    # Show integer percentage labels
    startangle=140,       # Rotate start for better spacing
    pctdistance=0.85,     # Position of percentage labels
    wedgeprops={'edgecolor': 'white'}  # White edge for a cleaner look
)


center_circle = plt.Circle((0, 0), 0.60, fc='white')
ax.add_artist(center_circle)

# title
ax.set_title("Do you romanticize the past?", fontsize=14)

# legend to the right of the donut
ax.legend(
    wedges,
    labels,
    title="Responses",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


art_list = []
for response in df["art_consumption[]"]:
    if response != "N/A":  # ignore missing entries
    
        items = [item.strip() for item in response.split(",")]
        art_list.extend(items)
        
#frequency Series for art consumption 
art_counts = pd.Series(art_list).value_counts()

# Process Cultural Attachment 

def classify_ca(resp):
    if pd.isnull(resp) or resp == "N/A":
        return "N/A"
    lower = resp.lower()

    if "yes" in lower:
        return "Yes"
   
    elif "no" in lower:
        return "No"

    elif "maybe" in lower:
        return "Maybe"
    else:
        return "Other"


ca_classified = df["cultural_attachment[]"].apply(classify_ca)
ca_counts = ca_classified.value_counts()

# Plotting Side-by-Side 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))


axes[0].barh(art_counts.index, art_counts.values, color="skyblue")
axes[0].set_title("Art Consumption Frequencies")
axes[0].set_xlabel("Count")
axes[0].invert_yaxis()


axes[1].bar(ca_counts.index, ca_counts.values, color="salmon", edgecolor="black")
axes[1].set_title(" Attachment to the culture from their childhood")
axes[1].set_xlabel("Response")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# The list of visual elements
visual_elements = [
    "Searose", 
    "Paul Delvaux and Giorgio de Chirico", 
    "Moody lighting, old furniture, trees drawn like a child",
    "Juxtaposition of smooth and real elements", 
    "Empty backgrounds feel like memories",
    "textures", 
    "3D and 2D elements clash", 
    "The playground lighting is nostalgic",
    "N/A", 
    "colours", 
    "Atmospheric lighting", 
    "The colour tones", 
    "Bright colours",
    "Early 3D rendering vibe", 
    "Colors and characters", 
    "Textured lines", 
    "Emptiness in backgrounds",
    "Round stone boulders and metal roofing", 
    "not sure", 
    "Lighting colors", 
    "Stark contrast",
    "Sharp angles like PlayStation games", 
    "Colours and composition"
]

# Create a pandas Series and count frequency for each response.
visual_series = pd.Series(visual_elements)
visual_counts = visual_series.value_counts()

# Create a colormap (using 'tab20' which can accommodate many categories)
cmap = plt.cm.get_cmap('tab20', len(visual_counts))
colors = [cmap(i) for i in range(len(visual_counts))]

# Create the pie chart (as a donut chart for a neater look)
plt.figure(figsize=(10, 10))
patches, texts, autotexts = plt.pie(
    visual_counts,
    labels=None,              # Labels will be shown in the legend
    autopct='%1.1f%%',        # Percentage labels on the chart
    startangle=140,           # Rotate for better layout
    colors=colors,
    pctdistance=0.85,         # Position of the percentage labels
    wedgeprops={'edgecolor': 'white'}
)

# Add a white circle in the center to achieve a donut effect
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

# Set title
plt.title("Distribution of Visual Elements Responses", fontsize=16)

# Create an external legend using the patches (color handles) and the labels from visual_counts.index.
plt.legend(
    patches,
    visual_counts.index,
    title="Visual Elements",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.axis('equal')  # Ensures the pie is drawn as a circle.
plt.tight_layout()
plt.show()

```
